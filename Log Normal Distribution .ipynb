```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(
    'income.csv', 
    index_col = None, 
    names = ['income', 'count'], 
    skiprows = 1
)

# "index_col" is set to None, 
# it means that no column from the input file will be 
# used as the index for the dataframe. 
# Instead, the dataframe will be assigned 
# a default integer index, starting from 0 and 
# incrementing for each row
```

```python
df.head()
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='income',y='count',data=df)

g.set_xticklabels(g.get_xticklabels(), 
                  rotation=45, 
                horizontalalignment='right');

# xtick labels -> x axis labels which is 45 deg rotated 
# g.set_xticklabels()
```

```python
sns.set(rc = {'figure.figsize' : (11.9, 9.2)})

g = sns.barplot(x = 'income', y = 'count', data = df)

g.set_xticklabels(g.get_xticklabels(),
                         rotation = 45, 
                         horizontalalignment = 'right')

g.set(xscale = 'log')
```
